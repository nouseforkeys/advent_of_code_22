```python
from matplotlib import pyplot as plt
import pandas as pd

from utils import load_from_file

data = pd.DataFrame({'depths': load_from_file('day_1.txt', cast_as=int)})

print(data)
```

```python
fig, ax = plt.subplots(1, 1)
data.plot(ax = ax, y='depths')
ax.grid('ON')
```

# Part 1
To do this, count the number of times a depth measurement increases from the previous measurement. (There is no measurement before the first measurement.)

```python
increase_counter = 0
for previous, this in zip(data['depths'][:-1], data['depths'][1:]):
    if this > previous:
        increase_counter += 1

print(f'The depth increases {increase_counter} times!')
```

# Part 2
Instead, consider sums of a three-measurement sliding window.

Your goal now is to count the number of times the sum of measurements in this sliding window increases from the previous sum.

```python
sum_increase_counter = 0
for a, d in zip(data['depths'][:-3], data['depths'][3:]):
    if d > a:
        sum_increase_counter += 1

print(
    'The rolling sum of depth with window size: 3 increases '
    f'{sum_increase_counter} times!'
)
```
